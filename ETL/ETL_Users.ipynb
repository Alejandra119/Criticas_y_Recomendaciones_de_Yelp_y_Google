### Importamos archivo

```python
user_url = "/content/drive/MyDrive/Final Project /Yelp/user.parquet"
```

```python
df_user = pd.read_parquet(user_url)
```

```python
df_user.info()
```

### 1. Seleccionamos columnas necesarias

```python
columnas = ['user_id', 'review_count', 'yelping_since', 'average_stars']
```

```python
df_user = df_user[columnas]
```

```python
df_user.info()
```

```python
df_user.sample(3)
```

### 2. Revisamos nulos y duplicados

```python
# Muestra la cantidad de valores nulos en cada columna
null_counts = df_user.isnull().sum()
print(null_counts)
```

```python
#Borramos duplicados
df_user['user_id'].drop_duplicates(inplace = True)
```

```python
df_user.info()
```

```python
df_user.sample(3)
```

### 3. Cambio de campo fecha

```python
df_user['yelping_since'] = pd.to_datetime(df_user['yelping_since'], format='%Y-%m-%d')
```
