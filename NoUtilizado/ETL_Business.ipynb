```python
import pandas as pd
import numpy as np
import fastparquet
```

### Importar archivo

```python
ruta = '/content/yelp_business.parquet'
df = pd.read_parquet(ruta)
df.info()
```

### 1. Filtrado por rubro "SteakHouses"

```python
#Creamos una mascara
mask = df['categories'].str.contains('Steakhouses',case = False) == True
df_steak = df[mask]
```

```python
df_steak.info()
```

```python
#Revisamos
df_steak.sample(3)
```

```python
#Reiniciamos Indices
df_steak.reset_index(drop=True, inplace=True)
#Revisamos
df_steak.head()
```

### 2. Filtramos las columnas necesarias

```python
print(df_steak.columns.tolist())
```

```python
#Definir las columnas que se usarán
columnas = ['business_id', 'name', 'address', 'city', 'latitude', 'longitude', 'stars', 'review_count', 'is_open', 'categories']
```

```python
#Filtramos
df_steak = df_steak[columnas]
```

```python
df_steak.head()
```

### 3. Tratamos nulos y duplicados

```python
# Muestra la cantidad de valores nulos en cada columna
null_counts = df_steak.isnull().sum()
print(null_counts)
```

```python
df_steak.drop_duplicates(inplace=True)
```

```python
df_steak.dropna(inplace=True)
```

```python
df_steak.info()
```

### 4. Tratamos datos

```python
df_steak.head()
```

- Reorganizamos y separamos columnas

```python
# Función lambda para reorganizar las categorías
reorganizar = lambda x: ', '.join(sorted(x.split(', '), key=lambda s: s != 'Steakhouses'))

# Aplicar la función lambda a la columna 'categories'
df_steak['categories'] = df_steak['categories'].apply(reorganizar)

# Separar las categorías en una nueva columna 'Categoria' y las demás en 'Subcategoria'
df_steak['categorie'] = df_steak['categories'].str.split(', ').str[0]
df_steak['subcategorie'] = df_steak['categories'].str.split(', ').str[1:].apply(', '.join)
```

```python
df_steak.head()
```

```python
# Eliminar la columna original 'categories'
df_steak.drop('categories', axis=1, inplace=True)
```

```python
df_steak.info()
```

```python
df_steak.sample(5)
```
